```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
taxi_data = pd.read_csv('combined_taxi_weather.csv')
```

```python
taxi_data.head()
```

```python
#dropping unnamed
taxi_data.drop('Unnamed: 0', inplace=True, axis= 1)
```

```python
# dropping columns which are not needed or have duplicates
taxi_data.drop(['date_y', 'hour_y', 'icon'], axis=1, inplace=True )
```

```python
taxi_data.sample(5)
```

```python
taxi_data.columns # Looking at the column names
```

```python
# Changing some column names

taxi_data.columns = ['vendorid', 'pickup_datetime', 'dropoff_datetime',
       'passenger_count', 'trip_distance', 'ratecodeid', 'store_and_fwd_flag',
       'pulocationid', 'dolocationid', 'payment_type', 'fare_amount', 'extra',
       'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge',
       'total_amount', 'congestion_surcharge', 'date', 'hour', 'key',
       'time', 'summary', 'precipIntensity', 'precipProbability', 'precipType',
       'temperature', 'apparentTemperature', 'dewPoint', 'humidity',
       'pressure', 'windSpeed', 'windGust', 'windBearing', 'cloudCover',
       'uvIndex', 'visibility', 'ozone', 'precipAccumulation']
```

```python
# There is some data from year 2008, 2009, 2020 and 2022 because of the merge
taxi_data['date'].value_counts()
```

```python
# Filtering the data so theres only data from 2021, but first I need to check the datatype and convert to datetime

print(taxi_data['pickup_datetime'].dtype) #object
taxi_data['pickup_datetime'] = pd.to_datetime(taxi_data['pickup_datetime'])
taxi_data['dropoff_datetime'] = pd.to_datetime(taxi_data['dropoff_datetime'])
print(taxi_data['pickup_datetime'].dtype) #datetime
```

```python
#Data for pickup between 2021 & 2022
taxi_data = taxi_data[(taxi_data['pickup_datetime'] > '2021') & (taxi_data['pickup_datetime'] < '2022')]
```

```python
#NANs to deal with
pd.DataFrame((taxi_data.isna().sum() / len(taxi_data)) * 100)
```

```python
#Looking at documentation, dropping these columns as they are not relevant
taxi_data.drop(['precipType','precipAccumulation' ], axis=1, inplace=True)
```

```python
# Dont need rows where passenger count is Null, 0 or greater than 10
taxi_data['passenger_count'].value_counts(dropna= False)
```

```python
#Filtering data that is not null in passenger_count
taxi_data = taxi_data[~taxi_data['passenger_count'].isna()] 

#Just want passenger count >0 and <10 as this is realistic
taxi_data = taxi_data[(taxi_data['passenger_count'] > 0) 
                      & (taxi_data['passenger_count'] < 10)]
```

```python
#Making a column for trip duration
taxi_data['trip_duration'] = taxi_data['dropoff_datetime']-taxi_data['pickup_datetime']
taxi_data.head()
```

```python
# Some have negative trip duration and there are some with max of 6 days

taxi_data['trip_duration'].dt.days.describe()
```

```python
#Checking trip duration > 0 days
taxi_data[taxi_data['trip_duration'].dt.days > 0]
```

```python
#Checking the negative trip durations these are errors in the data
taxi_data[taxi_data['trip_duration'].dt.days == -1]
```

```python
#Just want the data that is trips that can be completed < 1 day which = 0 days
taxi_data = taxi_data[taxi_data['trip_duration'].dt.days == 0]
```

```python
#No NANs now
taxi_data.isna().sum()
```

```python
# I see some weird trip distances more than 50 miles they seem like errors.

taxi_data['trip_distance'].value_counts()
```

```python
#Can see some negative values here - need to deal with them later
taxi_data.describe()
```

```python
# The distance to reach from the furthest end of new york city to other is end about 36 miles.  
# I am only looking at trips within new york city so 50 miles is a reasonable limit for a trip. 
taxi_data[taxi_data['trip_distance'] > 50]
```

```python
# Only need rows less than 50 miles
taxi_data = taxi_data[(taxi_data['trip_distance'] < 50)]
```

```python
taxi_data.shape
```

```python
#Looking at documention, dropping these as they are not relevant 
taxi_data.drop(['vendorid','ratecodeid', 'pulocationid', 
                'dolocationid', 'store_and_fwd_flag', 'payment_type' ], axis=1, inplace=True)
```

```python
taxi_data.columns
```

```python
taxi_data.shape
```

```python
#Looking at the descriptive stats above, some of these cols have negative data, I only need data >0 or >=0 depending on col
taxi_data = taxi_data[(taxi_data['fare_amount'] > 0) &
          (taxi_data['extra'] >= 0) &
          (taxi_data['tip_amount'] >= 0) &
          (taxi_data['mta_tax'] >= 0) &
          (taxi_data['tolls_amount'] >= 0) &
          (taxi_data['improvement_surcharge'] >= 0) &
          (taxi_data['total_amount'] > 0) &
          (taxi_data['congestion_surcharge'] >= 0)]
```

```python
taxi_data.shape
```

```python
#Making a col for tip percentage which is tip amount / total * 100
taxi_data['tip_percent'] = (taxi_data['tip_amount']/ taxi_data['total_amount']) * 100
```

```python
#The tip should be less than 100% of total, I did check and there wasnt any greater then 100%
taxi_data = taxi_data[taxi_data['tip_percent'] < 100]
```

```python
taxi_data.head()
```

```python
#Making a month column
taxi_data['month'] = taxi_data['pickup_datetime'].dt.month

# isocalendar starts from monday which is 1

#Making a col for day of week
taxi_data['day_of_week'] = taxi_data['pickup_datetime'].dt.isocalendar().day
```

```python
taxi_data.columns
```

```python
#Rearranging columns
taxi_data = taxi_data[['pickup_datetime', 'dropoff_datetime', 'trip_duration', 'day_of_week', 'month', 'hour',
                       'passenger_count', 'trip_distance', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 
       'tolls_amount', 'improvement_surcharge', 'congestion_surcharge', 'total_amount', 'tip_percent',
       'summary','temperature','humidity', 'windSpeed','uvIndex', 'visibility']].copy()
```

```python
#Resetting index
taxi_data.reset_index(inplace=True)
```

```python
#Dropping index
taxi_data.drop('index', inplace=True, axis=1)
```

```python
#Making a column for tip which has 0 for no tip and 1 for tip
taxi_data['tip_yes'] = np.where(taxi_data['tip_amount'] > 0, 1, 0)
```

```python
taxi_data.head()
```

```python
taxi_data.to_csv('final_data_cleaned.csv')
```
