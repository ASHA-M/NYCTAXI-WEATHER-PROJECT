```python
import requests as r
import pandas as pd
from getpass import getpass
```

```python
key = getpass() # API Key
```

```python
# this api gets the data by adding 4 parameters to the url, they are date, lat, long and key

date = '1609459200' #this date is unix date represting 1st jan 2021 obtained through #https://www.unixtimestamp.com
lat = '40.730610' # lat and long for new york city #https://www.latlong.net/place/new-york-city-ny-usa-1848.html
long = '-73.935242'
url = f'https://api.darksky.net/forecast/{key}/{lat},{long},{date}' #https://realpython.com/python-f-strings/
                                                                    #https://darksky.net/dev/docs
request = r.get(url)  # request and get the weather data for the date from URL
result = request.json() # turn the requested data into json datatype
result.keys()
```

```python
result['hourly'].keys()
```

```python
result['hourly']['data'] #I am only looking at hourly data from the requested json
```

```python
hourly = pd.json_normalize(result['hourly']['data']) # normalise and turn into DF
```

```python
hourly #for 1st Jan 2021
```

```python
# Converting Unix time into human readable date time
# the api actually gives us some rows before 1st Jan 2021

hourly['time'] = pd.to_datetime(hourly['time'],unit='s') # unit='s' helps with converting it to date time from unix
hourly
```

```python
# Need all days in year 2021
#start date (1st Jan 2021), stop date (up to 2nd Jan 2022, steps = 86400) #86400 seconds is 1 day
for date in range(1609459200, 1641085200, 86400): 
    print(date)
```

```python
# creating a loop to find the hourly weather data for all the days in 2021

#date = '1609459200' to '1641085200'
dfs = []
# getting the range of unix dates from 1st jan 2021 up to 2nd jan 2022, there are 86400 seconds in a day

for date in range(1609459200, 1641085200, 86400):  
    url = f'https://api.darksky.net/forecast/{key}/{lat},{long},{date}'

    request = r.get(url)
    result = request.json()
    hourly = pd.json_normalize(result['hourly']['data'])
    hourly['time'] = pd.to_datetime(hourly['time'],unit='s')
    dfs.append(hourly) #appending to dfs
```

```python
data = pd.concat(dfs, axis=0) #weather data for 2021
```
