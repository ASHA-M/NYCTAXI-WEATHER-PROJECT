```python
#Find out the last offset to get the last day
params = {'$limit': '50000',
          '$offset': '20650000', #start from this number (last day of 2021)
          '$$app_token' : token}
          
            
r = requests.request(method = 'GET',
                     url = 'https://data.cityofnewyork.us/resource/m6nq-qud6.json?vendorid=2',
                     params = params
                    )
json_data = r.json()
main_data = pd.json_normalize(json_data)
```

```python
main_data.tail()  # this is almost the last rows of data, 20650000/50000 = 413
```

```python
break # dont run the cell below as it will take a long time
```

```python
# Writing code here to add the rest of the data to the main_data

offset = 0
counter = 0
dfs = [] #to collect DF's

while counter < 414: #calculated above
    params = {'$limit': '50000',
                  '$offset': offset , #this will increase by 50,000 each time
                  '$$app_token' : token}


    r = requests.request(method = 'GET',
                             url = 'https://data.cityofnewyork.us/resource/m6nq-qud6.json?vendorid=2',
                             params = params
                            )
    json_data = r.json()
    temp_data = pd.json_normalize(json_data) # this is the data that contains the next 50000 rows
    temp_data['page'] = counter #create an extra column called page and sets values as same number as counter
    dfs.append(temp_data) #appending temp_data to dfs list, if it stops working I need to know at what point/page it stopped working
    offset = offset + 50000 #increase offset by 50000 each time
    print('current counter is ', counter) #print output of counter
    
    counter = counter + 1 #increase counter by 1 until 413

```

```python
taxi_data = pd.concat(dfs, axis=0) #concating all 413 small dfs into a large dataframe
```

# joining weather data to taxi data

```python
data['hour'] = data['hour'].astype('str')
data['date'] = data['date'].astype('str')
data['key'] = data['date'] + '-' + data['hour'] #key for taxi data
data.head()
```

```python
weather = pd.read_csv('final_weather.csv', index_col = 'Unnamed: 0')
```

```python
#https://pandas.pydata.org/docs/user_guide/merging.html
weather['hour'] = weather['hour'].astype('str')
weather['key'] = weather['date'] + '-' + weather['hour']
weather['key'].values[0] #unique key for weather data
```

```python
#left merge on key
df = pd.merge(data, weather, how='left', on='key') #get all data from taxi data and only the data from weather that matches using key
df.shape
```
